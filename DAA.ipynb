```python
# Pract2
import heapq

# Class to represent a node in the Huffman Tree
class MinHeapNode:
    def __init__(self, data, freq):
        self.data = data  # character
        self.freq = freq  # frequency of the character
        self.left = None  # left child
        self.right = None  # right child

    # Define comparison operators for priority queue (min-heap)
    def __lt__(self, other):
        return self.freq < other.freq

# Function to print the Huffman Codes
def print_codes(root, code=""):
    if root is None:
        return
    if root.data != '$':
        print(f"{root.data}: {code}")
    print_codes(root.left, code + "0")
    print_codes(root.right, code + "1")

# Function to build the Huffman Tree and print the codes
def huffman_code(data, freq):
    # Build the priority queue (min-heap)
    min_heap = [MinHeapNode(data[i], freq[i]) for i in range(len(data))]
    heapq.heapify(min_heap)

    # Iterate while the size of the heap is more than 1
    while len(min_heap) > 1:
        # Remove the two nodes of the lowest frequency
        left = heapq.heappop(min_heap)
        right = heapq.heappop(min_heap)

        # Create a new node with these two nodes as children and add back to heap
        merged = MinHeapNode('$', left.freq + right.freq)
        merged.left = left
        merged.right = right
        heapq.heappush(min_heap, merged)

    # Print the codes from the root of the Huffman Tree
    root = min_heap[0]
    print_codes(root)

# Test the Huffman Encoding
data = ['A', 'B', 'C', 'D']
freq = [23, 12, 34, 10]
huffman_code(data, freq)

```

```python
# Pract3

arr = [[500,30]]
w = 10
price = 0

arr = sorted(arr, key = lambda x : x[0] / x[1], reverse = True)

for i in range(len(arr)):
	itemWt = arr[i][1]
	itemP = arr[i][0]
	if(itemWt > w):
		price += w*(itemP / itemWt)
		break
	else:
		price += itemP
		w -= itemWt

print(price)

# Fractional Knapsack :
# Time Complexity: O(N * log N)
# Auxiliary Space: O(N)
```

```python
def knapsack(capacity, items, price, wt):
    # Initialize DP table with all values set to 0
    dp = [[0] * (capacity + 1) for _ in range(items + 1)]

    # Build the DP table
    for i in range(1, items + 1):
        for j in range(1, capacity + 1):
            if wt[i] <= j:
                # Choose the maximum of not taking or taking the current item
                dp[i][j] = max(dp[i - 1][j], price[i] + dp[i - 1][j - wt[i]])
            else:
                # If the item weight exceeds the current capacity, don't take it
                dp[i][j] = dp[i - 1][j]

    # The maximum profit for given capacity and items
    print(dp)
    return dp[items][capacity]

# Test data
capacity = 10
items = 4
price = [0, 3, 7, 2, 9]  # price of items (index 0 is a placeholder)
wt = [0, 2, 2, 4, 5]     # weight of items (index 0 is a placeholder)

max_profit = knapsack(capacity, items, price, wt)
print(f"Maximum Profit Earned: {max_profit}")
```

```python
def is_safe(board, row, col, n):
    # Check column for any other queen
    for i in range(row):
        if board[i][col] == 1:
            return False

    # Check upper left diagonal for any other queen
    i, j = row, col
    while i >= 0 and j >= 0:
        if board[i][j] == 1:
            return False
        i -= 1
        j -= 1

    # Check upper right diagonal for any other queen
    i, j = row, col
    while i >= 0 and j < n:
        if board[i][j] == 1:
            return False
        i -= 1
        j += 1

    return True

def print_board(board, n):
    for i in range(n):
        for j in range(n):
            if board[i][j] == 1:
                print("[Q]", end="")
            else:
                print("[]", end="")
        print()
    print("\n")

def n_queen(board, row, n):
    if row == n:
        print_board(board, n)
        return

    for col in range(n):
        if is_safe(board, row, col, n):
            board[row][col] = 1  # Place queen
            n_queen(board, row + 1, n)  # Recurse for next row
            board[row][col] = 0  # Backtrack

def solve_n_queens(n):
    # Initialize the board with all zeros
    board = [[0 for _ in range(n)] for _ in range(n)]
    n_queen(board, 0, n)
    print("--------All possible solutions--------")

# Input number of queens
n = int(input("Enter the number of queens: "))
solve_n_queens(n)
```

```python
import random

def partition(arr, p, r):
    x = arr[r]
    i = p - 1
    for j in range(p, r):
        if arr[j] <= x:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]  # Swap elements
    arr[i + 1], arr[r] = arr[r], arr[i + 1]
    return i + 1

def quick_sort(arr, p, r):
    if p < r:
        q = partition(arr, p, r)
        quick_sort(arr, p, q - 1)
        quick_sort(arr, q + 1, r)

# Randomized partitioning for QuickSort
def random_partition(arr, p, r):
    # Generate a random pivot index
    pivot_index = random.randint(p, r)
    arr[pivot_index], arr[r] = arr[r], arr[pivot_index]  # Swap pivot with end element

    return partition(arr, p, r)

def randomized_quick_sort(arr, p, r):
    if p < r:
        q = random_partition(arr, p, r)
        randomized_quick_sort(arr, p, q - 1)
        randomized_quick_sort(arr, q + 1, r)

# Test the randomized QuickSort
A = [23, 34, 54, 123, 34, 56, 67676, 112]
randomized_quick_sort(A, 0, len(A) - 1)
print("Sorted array:", A)
```
